```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
# Notice the data type pd.Categorical. 
# 
# Better clean up your data first and the you can while performaing analytics use this category . It saves lots of memory 
```

```python
tips['total_bill_str'] = tips['total_bill'].astype(str)
tips.head()
```

```python
tips.info()
```

```python
tips_sub_missing = tips.iloc[0:11, :]
```

```python
tips_sub_missing
```

```python
tips_sub_missing.loc[[1,3,5,7,9] , 'total_bill'] = 'missing'
```

```python
tips_sub_missing
```

```python
tips_sub_missing.info()
```

```python
# Both of below gives error because it can not convert string to integer 
#      tips_sub_missing['total_bill'].astype(float)
#       pd.to_numeric(tips_sub_missing['total_bill'])
```

```python
# You can see it just ignore the error , but data type is still object
pd.to_numeric(tips_sub_missing['total_bill'], errors='ignore')
```

```python
# This will force the convertsion to float , where ever it is not possible , it just gives fill that with NaN values.

pd.to_numeric(tips_sub_missing['total_bill'], errors='coerse')
```
