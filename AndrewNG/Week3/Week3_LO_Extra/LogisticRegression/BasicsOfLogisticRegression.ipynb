```python
import numpy as np  # Numerical computing
import matplotlib.pyplot as plt  # plotting core
```

```python
def g(z):
    return 1/(1 + np.exp(-z))
```

```python
def test_g():
    z = np.linspace(-10, 10, 100)
    plt.plot(z, g(z))
    plt.grid(True)
    plt.xlabel("z")
    plt.ylabel("Sigmoid(z)")
    plt.annotate('if z < .5 , then y = 0', (-10, 0.2), fontsize=12)
    plt.annotate('if z => .5 , then y = 1', (3, 0.8), fontsize=12)
    plt.title("Sigmoid Function g(z) = 1/(1 + exp(-z))", fontsize=12, color="blue")
    plt.show()
test_g()
```

```python
def plot_cost_function():
    z = np.linspace(-10, 10, 100)
    plt.plot(g(z), -np.log(g(z)))
    plt.plot(g(z), -np.log(1-g(z)))
    plt.xlabel("g(z)", fontsize=12, color="blue")
    plt.ylabel("cost", fontsize=12, color="blue")
    plt.annotate('- np.log(g(z))', (0, 10), fontsize=12, color="blue")
    plt.annotate("- np.log(1 - g(z))", (.8, 10), fontsize=12, color="blue")
    plt.grid(True)
    plt.show()
plot_cost_function()
```
